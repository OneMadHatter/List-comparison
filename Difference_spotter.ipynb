```python
import numpy as np
import scipy as sp
import pandas as pd
```

```python
old_items=pd.read_csv("Prova1", header=None, engine='python')
new_items=pd.read_csv("Prova 2", header=None, engine='python')
```

```python
before=list(old_items[0].to_numpy())
after=list(new_items[0].to_numpy())
matches=set.intersection(set(before),set(after))
intersection_list = list(matches)
```

```python
def compare_new(new_items, matches):
    for item in new_items:
        if item not in matches:
            print(item, "is now in the list")
        else:
            pass

compare_new(after, matches)
```

```python
def compare_old(old_items, matches):
    for item in old_items:
        if item not in matches:
            print(item, "is not in the list anymore.")
        else:
            pass

compare_old(before, matches)
```
